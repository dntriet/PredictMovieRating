```python
import pandas as pd

# Load the dataset
file_path = 'D:\Tai_lieu_SV\Hk6\IS252-Khai_thac_du_lieu\DoAn\TranPhuongLam_21521059_21522134_2152711_21522751_Final_report\cleaned_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head(), data.describe(), data.info()
```

```python
import matplotlib.pyplot as plt

# Biểu đồ phân bố năm phát hành
plt.figure(figsize=(10, 6))
plt.hist(data['startYear'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phân bố năm phát hành')
plt.xlabel('Năm phát hành')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.show()

# Biểu đồ phân bố thời lượng phim
plt.figure(figsize=(10, 6))
plt.hist(data['runtimeMinutes'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phân bố thời lượng phim')
plt.xlabel('Thời lượng phim (phút)')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.show()

# Biểu đồ phân bố điểm đánh giá trung bình
plt.figure(figsize=(10, 6))
plt.hist(data['averageRating'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phân bố điểm đánh giá trung bình')
plt.xlabel('Điểm đánh giá trung bình')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.show()

# Biểu đồ số lượng thể loại phim
plt.figure(figsize=(10, 6))
plt.hist(data['numGenres'], bins=range(1, 5), edgecolor='k', alpha=0.7, align='left')
plt.title('Số lượng thể loại phim')
plt.xlabel('Số lượng thể loại')
plt.ylabel('Số lượng phim')
plt.grid(True)
plt.xticks(range(1, 4))
plt.show()

# Biểu đồ phân bố số lượng đánh giá
plt.figure(figsize=(10, 6))
plt.hist(data['numVotes'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phân bố số lượng đánh giá')
plt.xlabel('Số lượng đánh giá')
plt.ylabel('Số lượng phim')
plt.yscale('log')  # Sử dụng log scale để trực quan hóa tốt hơn
plt.grid(True)
plt.show()
```
