```python
import pandas as pd

# Load the dataset
file_path = 'D:\Tai_lieu_SV\Hk6\IS252-Khai_thac_du_lieu\DoAn\TranPhuongLam_21521059_21522134_2152711_21522751_Final_report\cleaned_data.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head(), data.describe(), data.info()
```

```python
import matplotlib.pyplot as plt

# Biu  phn b nm pht hnh
plt.figure(figsize=(10, 6))
plt.hist(data['startYear'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phn b nm pht hnh')
plt.xlabel('Nm pht hnh')
plt.ylabel('S lng phim')
plt.grid(True)
plt.show()

# Biu  phn b thi lng phim
plt.figure(figsize=(10, 6))
plt.hist(data['runtimeMinutes'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phn b thi lng phim')
plt.xlabel('Thi lng phim (pht)')
plt.ylabel('S lng phim')
plt.grid(True)
plt.show()

# Biu  phn b im nh gi trung bnh
plt.figure(figsize=(10, 6))
plt.hist(data['averageRating'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phn b im nh gi trung bnh')
plt.xlabel('im nh gi trung bnh')
plt.ylabel('S lng phim')
plt.grid(True)
plt.show()

# Biu  s lng th loi phim
plt.figure(figsize=(10, 6))
plt.hist(data['numGenres'], bins=range(1, 5), edgecolor='k', alpha=0.7, align='left')
plt.title('S lng th loi phim')
plt.xlabel('S lng th loi')
plt.ylabel('S lng phim')
plt.grid(True)
plt.xticks(range(1, 4))
plt.show()

# Biu  phn b s lng nh gi
plt.figure(figsize=(10, 6))
plt.hist(data['numVotes'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Phn b s lng nh gi')
plt.xlabel('S lng nh gi')
plt.ylabel('S lng phim')
plt.yscale('log')  # S dng log scale  trc quan ha tt hn
plt.grid(True)
plt.show()
```
